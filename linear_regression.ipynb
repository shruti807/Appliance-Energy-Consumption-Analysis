```python
import pandas as pd
```

```python
df = pd.read_excel('/content/data_application_energy.xlsx')
```

```python
df
```

```python
# prompt: give me python code for linear regression plot of above data between colummns  rv1 and rv2 lights	T1

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming 'rv1', 'rv2', 'lights', and 'T1' are column names in your DataFrame 'df'

X = df[['rv1']]  # Independent variable (rv1)
y = df['T1']  # Dependent variable (T1)

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('rv1')
plt.ylabel('T1')
plt.title('Linear Regression of rv1 vs T1')
plt.legend()
plt.show()
```
